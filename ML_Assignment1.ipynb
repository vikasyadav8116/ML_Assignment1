```python
import numpy as np
import matplotlib.pyplot as plt

# Load your dataset (replace with your actual file paths)
X = np.loadtxt('/content/linearX.csv')  # Replace with your file path
y = np.loadtxt('/content/linearY.csv')  # Replace with your file path

# Add a column of ones to X for the intercept term (theta_0)
X = np.c_[np.ones(X.shape[0]), X]

# Initialize theta (parameters)
theta_initial = np.zeros(X.shape[1])

# Number of training examples
m = len(y)

# Cost function
def compute_cost(X, y, theta):
    predictions = X.dot(theta)
    cost = (1 / (2 * m)) * np.sum(np.square(predictions - y))
    return cost

# Batch Gradient Descent
def batch_gradient_descent(X, y, theta, alpha, iterations):
    cost_history = np.zeros(iterations)
    for i in range(iterations):
        predictions = X.dot(theta)
        theta = theta - (alpha / m) * X.T.dot(predictions - y)
        cost_history[i] = compute_cost(X, y, theta)

        # Convergence criteria (stopping when the change in cost is very small)
        if i > 0 and abs(cost_history[i] - cost_history[i-1]) < 1e-6:
            break
    return theta, cost_history

# Stochastic Gradient Descent (SGD)
def stochastic_gradient_descent(X, y, theta, alpha, iterations):
    cost_history = []
    for i in range(iterations):
        # Shuffle the data
        data = np.c_[X, y]
        np.random.shuffle(data)
        X_shuffled = data[:, :-1]
        y_shuffled = data[:, -1]

        for j in range(m):
            x_j = X_shuffled[j, :].reshape(1, -1)
            y_j = y_shuffled[j]
            prediction = np.dot(x_j, theta)
            gradient = (prediction - y_j) * x_j
            theta = theta - alpha * gradient.T

        cost_history.append(compute_cost(X, y, theta))

        # Convergence criteria (stopping when the change in cost is very small)
        if i > 0 and abs(cost_history[-1] - cost_history[-2]) < 1e-6:
            break
    return theta, cost_history

# Mini-Batch Gradient Descent
def mini_batch_gradient_descent(X, y, theta, alpha, iterations, batch_size=32):
    cost_history = []
    for i in range(iterations):
        # Shuffle the data
        data = np.c_[X, y]
        np.random.shuffle(data)
        X_shuffled = data[:, :-1]
        y_shuffled = data[:, -1]

        for j in range(0, m, batch_size):
            X_batch = X_shuffled[j:j+batch_size]
            y_batch = y_shuffled[j:j+batch_size]
            predictions = X_batch.dot(theta)
            gradient = (1 / batch_size) * X_batch.T.dot(predictions - y_batch)
            theta = theta - alpha * gradient

        cost_history.append(compute_cost(X, y, theta))

        # Convergence criteria (stopping when the change in cost is very small)
        if i > 0 and abs(cost_history[-1] - cost_history[-2]) < 1e-6:
            break
    return theta, cost_history

# Set a suitable learning rate
learning_rate = 0.5
iterations = 1000

# Run Gradient Descent for each type

# Batch Gradient Descent
theta_batch, cost_batch = batch_gradient_descent(X, y, theta_initial.copy(), learning_rate, iterations)

# Stochastic Gradient Descent
theta_sgd, cost_sgd = stochastic_gradient_descent(X, y, theta_initial.copy(), learning_rate, iterations)

# Mini-Batch Gradient Descent
theta_mini_batch, cost_mini_batch = mini_batch_gradient_descent(X, y, theta_initial.copy(), learning_rate, iterations, batch_size=32)

# Plot the Cost Function vs Iterations
plt.figure(figsize=(10, 6))
plt.plot(cost_batch, label="Batch Gradient Descent", color='blue')
plt.plot(cost_sgd, label="Stochastic Gradient Descent", color='red')
plt.plot(cost_mini_batch, label="Mini-Batch Gradient Descent", color='green')
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Cost Function vs Iterations for Different Gradient Descent Methods")
plt.legend()
plt.grid(True)
plt.show()

# Output the results
print(f"Batch GD final cost: {cost_batch[-1]}")
print(f"SGD final cost: {cost_sgd[-1]}")
print(f"Mini-Batch GD final cost: {cost_mini_batch[-1]}")
```
