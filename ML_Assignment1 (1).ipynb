```python
import numpy as np
import matplotlib.pyplot as plt

# Load your dataset (replace with your actual file paths)
X = np.loadtxt('/content/linearX.csv')  # Replace with your file path
y = np.loadtxt('/content/linearX.csv')  # Replace with your file path

# Add a column of ones to X for the intercept term (theta_0)
X = np.c_[np.ones(X.shape[0]), X]

# Initialize theta (parameters)
theta = np.zeros(X.shape[1])

# Define hyperparameters
alpha = 0.5  # Learning rate
iterations = 1000
m = len(y)  # Number of training examples

# Cost function
def compute_cost(X, y, theta):
    predictions = X.dot(theta)
    cost = (1 / (2 * m)) * np.sum(np.square(predictions - y))
    return cost

# Gradient Descent
def gradient_descent(X, y, theta, alpha, iterations):
    cost_history = np.zeros(iterations)
    for i in range(iterations):
        predictions = X.dot(theta)
        theta = theta - (alpha / m) * X.T.dot(predictions - y)
        cost_history[i] = compute_cost(X, y, theta)

        # Convergence criteria (stopping when the change in cost is very small)
        if i > 0 and abs(cost_history[i] - cost_history[i-1]) < 1e-6:
            break

    return theta, cost_history

# Run Gradient Descent
theta, cost_history = gradient_descent(X, y, theta, alpha, iterations)

# Output the results
print(f"Converged theta: {theta}")
print(f"Final cost: {cost_history[-1]}")

# Plot the cost history
plt.plot(range(len(cost_history)), cost_history, label="Cost Function")
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Convergence of Gradient Descent")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load datasets
# Replace these file paths with your actual dataset paths
X = np.loadtxt('/content/linearX.csv')  # Independent variable file
y = np.loadtxt('/content/linearY.csv')    # Dependent variable file

# Check shapes of the datasets
print("Original X shape:", X.shape)
print("Original y shape:", y.shape)

# Align data if shapes are mismatched
if X.shape[0] != y.shape[0]:
    print("Mismatch in rows! Aligning data...")
    min_rows = min(X.shape[0], y.shape[0])
    X = X[:min_rows]  # Trim X to match
    y = y[:min_rows]  # Trim y to match

# Final shape check
print("Aligned X shape:", X.shape)
print("Aligned y shape:", y.shape)

# Add a column of ones to X for the intercept term (theta_0)
X = np.c_[np.ones(X.shape[0]), X]  # Shape becomes (m, n+1)

# Initialize parameters (theta) and settings
theta = np.zeros(X.shape[1])  # Initialize theta with zeros
learning_rate = 0.1           # Learning rate
iterations = 100              # Number of iterations

# Cost function
def compute_cost(X, y, theta):
    m = len(y)  # Number of training examples
    predictions = X.dot(theta)  # Predictions using current theta
    cost = (1 / (2 * m)) * np.sum(np.square(predictions - y))  # Compute cost
    return cost

# Gradient descent function
def gradient_descent(X, y, theta, learning_rate, iterations):
    m = len(y)
    cost_history = []  # To store cost at each iteration

    for _ in range(iterations):
        predictions = X.dot(theta)  # Compute predictions
        gradient = (1 / m) * X.T.dot(predictions - y)  # Compute gradient
        theta = theta - learning_rate * gradient  # Update theta
        cost = compute_cost(X, y, theta)  # Compute cost
        cost_history.append(cost)  # Save cost

    return theta, cost_history

# Run gradient descent
theta_final, cost_history = gradient_descent(X, y, theta, learning_rate, iterations)

# Output final theta and cost
print("Final theta (parameters):", theta_final)
print("Final cost:", cost_history[-1])

# Plot Cost Function vs Iterations
plt.figure(figsize=(10, 6))
plt.plot(range(len(cost_history)), cost_history, color='blue', label="Cost")
plt.title("Cost Function vs Iterations")
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.legend()
plt.grid(True)
plt.show()

# Plot the data and the fitted line
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 1], y, color='red', label="Data")  # Plot the original data
plt.plot(X[:, 1], X.dot(theta_final), color='blue', label="Fitted Line")  # Plot the fitted line
plt.title("Data and Fitted Line")
plt.xlabel("Independent Variable")
plt.ylabel("Dependent Variable")
plt.legend()
plt.grid(True)
plt.show()
```
